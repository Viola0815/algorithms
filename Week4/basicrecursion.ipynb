```python
############################################################
# Recursion.py 
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2021
###########################################################

############################################################
#  All imports here
###########################################################

############################################################
#  class  test factorial
###########################################################    
class Recursion():
    def __init__(self):
        #Nothing can be added here
        pass

    ############################################################
    #          WRITE ALL YOUR PUBLIC FUNCTION BELOW
    ########################################################### 
    def factI(self, n):
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

    def factR(self, n):
        if n == 0 or n == 1:
            return 1
        return n * self.factR(n - 1)

    def reverseI(self, n):
        reversed_num = 0
        while n > 0:
            digit = n % 10
            reversed_num = reversed_num * 10 + digit
            n //= 10
        return reversed_num
    
    def reverseR(self, n):
        if n < 10:
            return n
        else:
            return (n % 10) * 10**(len(str(n)) - 1) + self.reverseR(n // 10)

    def printI(self, n, reverse):
        for digit in str(n)[::-1] if reverse else str(n):
            print(digit)

    def printR(self, n, reverse):
        if n < 10:
            print(n)
        else:
            if reverse:
                print(n % 10, end=" ")
                self.printR(n // 10, reverse)
            else:
                self.printR(n // 10, reverse)
                print(n % 10, end=" ")

    ############################################################
    #          WRITE ALL YOUR PRIVATE FUNCTION BELOW
    ########################################################### 
   
```
